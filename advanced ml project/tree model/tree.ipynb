```python
def bar_chart(features):
    survived = d_train[d_train['Survived']==1][features].value_counts()
    dead = d_train[d_train['Survived']==0][features].value_counts()
    df = pd.DataFrame([survived,dead])
    df.index= ['survived','dead']
    df.plot(kind = 'bar',stacked = True,figsize=(10,5))
```
